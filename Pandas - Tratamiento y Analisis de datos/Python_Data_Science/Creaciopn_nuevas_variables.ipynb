```python
import pandas as pd
```

<font color= yellow> Creando Nuevas Variables

```python
datos = pd.read_csv('data/alquiler_residencial.csv', sep=';')
datos.head(10)
```

```python
# Crear nueva varibale o columna
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos
```

```python
# Calcular el valor por metro cuadrado y rendeado a dos decimales
datos['Valor m2'] = ((datos['Valor']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
# Convertir las variables 'Casa en condominio' y 'Casa de villa' en 'Casa'
Casa = ['Casa', 'Casa en condominio', 'Casa de villa']
datos['Tipo agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in Casa else 'Departamento')
datos.head(10)
```

<font color= yellow> Excluyendo o eliminando Variables

```python
# Eliminar las columnas 'Valor bruto', 'Valor m2' y 'Tipo agrupado'
# Crear un DataFrame de ejemplo con las columnas que queremos eliminar
datos_aux = pd.DataFrame(datos[['Impuesto', 'Valor Bruto', 'Valor m2', 'Tipo agrupado']])
datos_aux.head(10)
```

```python
# Primer metodo para eliminar una columna
del datos_aux['Valor Bruto']
datos_aux.head(10)
```

```python
# Segundo metodo para eliminar una columna
datos_aux.pop('Valor m2')
datos_aux.head(10)
```

```python
# Tercer metodo para eliminar una columna
# EL metodo drop por defecto elimina filas, por eso se debe poner "axis= 1"
datos_aux.drop(['Tipo agrupado'], axis= 1, inplace= True)
datos_aux.head(10)
```

```python
# Elimnar las columnas que creamos en el dataframe datos
datos.drop(['Valor Bruto', 'Valor m2'], axis= 1, inplace= True)
```

```python
datos.head(10)
```

```python
# Guardar o sobreescribir el archivo modificado
datos.to_csv('data/alquiler_residencial.csv', sep=';', index= False)
```
