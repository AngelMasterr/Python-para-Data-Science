```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('data/alquiler_residencial_sin_outliers.csv', sep=';')
datos.head(10)
```

Creando la configuracion de los graficos

```python
# Crear el area default del grafico
area = plt.figure(figsize=(16, 12))
```

```python
# Dividir el grafico en 2 filas y 2 columnas y darle una posicion "g" a cada grafico
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
# Crear grafico de dispersion
g1.scatter(datos['Valor'], datos['Area'])
g1.set_title('Valor por Area')

# Crear grafico de histograma
g2.hist(datos['Valor'])
g2.set_title('Histograma')

# Crear grafico con muestras de datos random con 100 registros
datos_g3 = datos['Valor'].sample(100)
datos_g3.index = range(datos_g3.shape[0])
g3.plot(datos_g3)
g3.set_title('Muestra (Valor)')

# Crear grafico clasificando segun el Tipo de la columna Valor
grupo = datos.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Medio por Tipo')
```

```python
# Guardar el grafico
area.savefig('imagenes/grafico.png', dpi = 300, bbox_inches = 'tight')
```
