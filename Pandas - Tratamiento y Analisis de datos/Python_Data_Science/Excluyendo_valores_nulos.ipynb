<font color= yellow> Excluyendo valores nulos

```python
datos = pd.read_csv('data/alquiler_residencial.csv', sep=';')
datos.head(10)
```

```python
# Devuelve True en los valores que aparezcan nullos o NaN
datos.isnull()
```

```python
# Devuelve False en los valores que aparezcan nullos o NaN
datos.notnull()
```

```python
datos.info()
```

```python
# Mostra en la columna 'Valor' los valores que tengan NaN 'Null'
datos[datos['Valor'].isnull()]
```

```python
# Eliminar las filas que contengan el valor NaN de la columna 'Valor'
datos.dropna(subset= ['Valor'], inplace= True)
datos
```

<font color= yellow> Excluir valores nulos de otra manera

```python
datos
```

Eliminar las filas de Tipo: 'Departamento' siempre y cuando el Mantenimiento sea 'NaN'

```python
# Conocer cuantas filas en 'Mantenimiento' tienen 'NaN'
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
# Seleccionar las columnas que cumplan la condicion que necesitamos
seleccion = (datos['Tipo'] == 'Departamento') & (datos['Mantenimiento'].isnull())
seleccion
```

```python
# Eliminar las filas de Tipo: 'Departamento' y el Mantenimiento es 'NaN'
# se pone '~' para traer lo contrario del booleano, en este caso la seleccion da True a los valores que queremos eliminar
# por eso se pone '~' para traer lo opuesto
datos = datos[~seleccion]
datos
```

Cambiar los valores nulos NaN de las columnas 'Mantenimiento' y 'Impuesto' por 0

```python
datos = datos.fillna({'Mantenimiento':0, 'Impuesto': 0})
datos
```

```python
datos.info()
```

```python
# Exportar el archivo sobreescribiendo el que ye tenemos pero ahora sin los valores de NaN
datos.to_csv('data/alquiler_residencial.csv', sep=';', index= False)
```
