```python
import pandas as pd
```

### <font color= yellow> Selecciones y Frecuencias

```python
datos = pd.read_csv('data/alquiler_residencial.csv', sep=';')
datos.head(10)
```

Seleccione solo lo inmuebles clasificados con Tipo 'Departamento'

```python
# Otra manera de selccion: 
# seleccion_dep = datos['Tipo'] == 'Departamento' 
seleccion_dep = datos['Tipo'] .isin(['Departamento']) 
seleccion_dep
```

```python
datos_dep = datos[seleccion_dep]
datos_dep
```

```python
# Cuantas filas hay con el Tipo 'Departamento'
datos_dep.shape[0]
```

Seleccione solo los inmuebles clasificados con Tipo 'Casa', 'Casa en condominio' y 'Casa de villa' 

```python
# Otra manera de selccion: 
# seleccion_dep = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
seleccion_casa_cond_villa = datos['Tipo'].isin(['Casa', 'Casa en condominio', 'Casa de villa'])
seleccion_casa_cond_villa
```

```python
datos_casa_cond_villa = datos[seleccion_casa_cond_villa]
datos_casa_cond_villa
```

```python
# Cuantas filas cumplen con este criterio
datos_casa_cond_villa.shape[0]
```

Seleccione lo inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites

```python
datos
```

```python
# Otra manera
# seleccion_area = (datos['Area'] >= 60) & (datos['Area'] <= 100)
seleccion_area = datos['Area'].between(60,100)
seleccion_area
```

```python
datos_area = datos[seleccion_area]
datos_area
```

```python
# Cuantos datos cumplen con este criterio
datos_area.shape[0]
```

Selecciones lo inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2000.00

```python
datos
```

```python
seleccion_cuartos_alq = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
seleccion_cuartos_alq
```

```python
# Cuantas filas cumplen este criterio
datos_cuartos_alq = datos[seleccion_cuartos_alq]
datos_cuartos_alq.shape[0]
```

```python
print(f"Numero de inmuebles clasificados con Tipo 'Departamento': {datos_dep.shape[0]}")
print(f"Numero de inmuebles clasificados con Tipo 'Casa', 'Casa en condominio' y 'Casa de villa': {datos_casa_cond_villa.shape[0]}")
print(f"Numero de inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites: {datos_area.shape[0]}")
print(f"Numero de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2000.00: {datos_cuartos_alq.shape[0]}")
```
