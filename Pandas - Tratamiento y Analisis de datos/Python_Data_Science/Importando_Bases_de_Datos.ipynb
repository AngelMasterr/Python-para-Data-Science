```python
import pandas as pd
```

### Importando un archivo .json

```python
json = open('extras/alquiler.json')
print(json.read())
```

### Abrir un archivo .json en un DataFrame de pandas

```python
df_json = pd.read_json('extras/alquiler.json')
df_json
```

### Importando un archivo .txt

```python
txt = open('extras/alquiler.txt')
print(txt.read())
```

### Abrir un archivo .txt en un DataFrame de pandas

```python
df_txt = pd.read_table('extras/alquiler.txt')
df_txt
```

### Abrir un archivo .xlsx en un DataFrame de pandas

```python
df_xlsx = pd.read_excel('extras/alquiler.xlsx')
df_xlsx
```
