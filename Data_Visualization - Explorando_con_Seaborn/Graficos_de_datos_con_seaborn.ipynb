<font color= yellow> Test de hipotesis

```python
# scatter_kws={'s': 10}: tamaño de los puntos del grafico
sns.lmplot(data=tarjetas, x='iu', y='valor_factura',  height=3, aspect=1.2, col='moroso', scatter_kws={'s': 10});
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribución de los grupos morosos y no morosos es la misma

H<sup>alt<sup>

La distribución de los grupos morosos y no morosos NO es la misma

```python
# crear los grupos morosos y no morosos
moroso = tarjetas.query("moroso == 1").valor_factura
moroso
```

```python
no_moroso = tarjetas.query("moroso == 0").valor_factura
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print(f"El p_value de nuestro test es {resultado.pvalue}")
```
